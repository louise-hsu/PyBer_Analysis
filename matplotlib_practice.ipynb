```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# skill drill- object oriented
fig, ax = plt.subplots()
# change line color to gree, marker to diamond, line width to 2, leged for Boston
ax.plot(x_axis, y_axis, label = "Boston", color = "green", linewidth = 2, marker = "d")
# add a legend
ax.legend()
# label axes and title
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_title("PyBer Fare by Month")
# set y_axis limit
ax.set_ylim(0,45)
# add grid lines
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# if use plt.plot() function to make a scatter plot, you need to add a lower case "o" as a parameter
plt.plot(x_axis, y_axis, 'o')
```

```python
# create scatter ploy using the plt.scatter()
plt.scatter(x_axis, y_axis)
```

```python
# Skill Drill - MATLAB scatter plot
# red markers, switch axis
plt.scatter(y_axis, x_axis, c = "red", label = "Chicago")
# add legend
plt.legend()
# add a title and axes label
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
# invert the y_axis - jan top
plt.gca().invert_yaxis()
# add limit for x_axis
plt.xlim(0,50)
```

```python
# bubble chart
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# adjust the size 
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# skill Drill - bubble chart object oriented
fig, ax = plt.subplots()
#switch the axis, skyblue, black edge, 20% transparent, 5 x marker size, linewidth = 2
ax.scatter(y_axis, x_axis, color = "skyblue", alpha = .80, edgecolors = "black", linewidth = 2, label = "Boston", s = [ i * 5 for i in y_axis])
# add title and axes
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
# add legend
ax.legend()
# invert y_axis
ax. invert_yaxis()
# limit for x_axis
ax.set_xlim(0,50)
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Skill drill - object-oriented approach pie chart
# new colors
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# explode the two highest values
explode_values = (0, 0, 0.2 , 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# increase fig size to 8x8, %, colors, explode, shadow
fig, ax = plt.subplots(figsize =(8,8))
ax.pie(y_axis,
       labels= x_axis, 
       autopct='%.1f%%', 
       labeldistance = 1.2,
       radius = 1,
       colors = colors, 
       explode = explode_values, 
       shadow = True,
       startangle = 85
      )
#reverse order
ax.invert_xaxis()


plt.show()
```
